```python
import requests, bs4
```

```python
url = "https://www.apimmigrations.com/client-review/"

req = requests.get(url)

soup = bs4.BeautifulSoup(req.text, "html5lib")

soup
```

```python
reviews = soup.select(".post-contentnew")

for i in range(len(reviews)):
    para = str(reviews[i])
    para = para.split("<p>")
    para[1] = para[1].split("</p>")
    print(i)
    print(para[1][0])
    print("")
```
